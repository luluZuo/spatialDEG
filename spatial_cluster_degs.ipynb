```python
XYmap = pd.DataFrame({'x': x, "y": y})
XYnbrs = NearestNeighbors(n_neighbors=20, algorithm='auto', metric='euclidean').fit(XYmap)
_, XYindices = XYnbrs.kneighbors(XYmap)
nbr_id = XYindices[test_df.index]
nbr_id
```

```python
## neighbor count
nbr_id_unique = np.unique(nbr_id)
group_id = []
for x in np.nditer(nbr_id_unique):
    group_id.append(df.loc[x, 'group'])
nbr_group = Counter(group_id)
nbr_group
```

```python
## ratio
group=[0,1,2,3,4,5,6]
group_num = dict()
ratio = dict()
for i in group:
    group_num[i] = df['group'].value_counts()[i]
    ratio[i] = nbr_group[i]/group_num[i]
nbr_group
ratio
```

```python
test_group=1
ratio.values()
result = [i for i, e in enumerate(ratio.values()) if e > 0.5]
result.remove(test_group)
result
```
